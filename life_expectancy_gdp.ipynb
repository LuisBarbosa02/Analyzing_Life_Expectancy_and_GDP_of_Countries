# Analyzing the Life Expectancy and GDP of Countries

Data gathered from the World Health Organization and World Bank.

#### Importing necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Importing the data and analyzing it's features:

```python
df = pd.read_csv('all_data.csv')

df.rename(columns={'Life expectancy at birth (years)': 'Life expectancy'}, inplace=True)

df.info()
df.sample(n=7)
```

This dataset consists of:
- Name of the country;
- Year the data was gathered;
- The life expectancy in that year;
- The GDP in that year.

As you can see, even though all the data that we'll work with is within this dataset, just looking at data in a table doesn't give us much information on what we're looking for.

## Using graphs to analyze the dataset

Now we'll use graphs to look for insights that we can't see just looking in data tables.

#### Analyzing the GDP of countries over the years:

```python
plt.figure(figsize=(7.4, 5.8))

sns.lineplot(data=df, x='Year', y='GDP', hue='Country')

plt.legend(loc=(1.02, .5))
plt.ylabel('GDP (in trillion)', fontsize=13)
plt.xlabel('Year', fontsize=13)
plt.title('GDP of Countries over Years', fontsize=15)


plt.show()
plt.clf()

print('\n')
print('Especifying the GDP of each country over the years:')
print('\n')

sns.FacetGrid(data=df, col='Country', col_wrap=3, hue='Country', sharey=False).map(sns.lineplot, 'Year', 'GDP').add_legend().set_axis_labels('Year', 'GDP (in U.S. dollars)')

plt.show()
plt.clf()
```

So, the tendency is that the countries increases their GDP over the years.

#### Analyzing the life expectancy of countries over the years:

```python
plt.figure(figsize=(7.4, 5.8))

sns.lineplot(data=df, x='Year', y='Life expectancy', hue='Country')

plt.xlabel('Year', fontsize=13)
plt.ylabel('Life Expectancy (in years)', fontsize=13)
plt.title('Life Expectancy of Countries over Years', fontsize=15)

plt.show()
plt.clf()

print('\n')
print('Especifying the life expectancy of each country over the years:')
print('\n')

sns.FacetGrid(data=df, col='Country', col_wrap=3, hue='Country', sharey=False).map(sns.lineplot, 'Year', 'Life expectancy').add_legend().set_axis_labels('Year', 'Life Expectancy (in years)')

plt.show()
plt.clf()
```

As the GDP, the life expectancy increases over the years.

#### Checking the correlation between life expectancy and GDP:

```python
plt.figure(figsize=(7.4, 5.8))

sns.scatterplot(data=df, x='Life expectancy',  y='GDP', hue='Country')

plt.xlabel('Life Expectancy (in years)', fontsize=13)
plt.ylabel('GDP (in trillions)', fontsize=13)
plt.title('Correlation Between GDP and Life Expectancy', fontsize=15)

plt.show()
plt.clf()

print('\n')
print('Especifying the correlation between GDP and life expectancy for each country:')
print('\n')

sns.FacetGrid(data=df, col='Country', hue='Country', col_wrap=3, sharey=False).map(sns.scatterplot, 'Life expectancy', 'GDP').add_legend().set_axis_labels('Life expectancy (in years)', 'GDP')

plt.show()
plt.clf()
```

The correlation between GDP and life expectancy is stronger when the country is poor. As the country get richer, the GDP starts to not affect the life expectancy that much, even though still affects it.
